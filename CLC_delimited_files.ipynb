```python
#sample1=pd.read_csv(r'C:\Users\sselvaku\OneDrive - Systech Solutions, Inc\Dev\Sample_1.txt', sep='|')
sample1=pd.read_csv(r'C:\Users\sselvaku\OneDrive - Systech Solutions, Inc\Dev\Sample_1_mismatch.txt', sep='|')
#sample2=pd.read_csv(r'C:\Users\sselvaku\OneDrive - Systech Solutions, Inc\Dev\Sample_2.txt', sep='|')
sample2=pd.read_csv(r'C:\Users\sselvaku\OneDrive - Systech Solutions, Inc\Dev\Sample_2_mismatch.txt', sep='|')
```

```python
col1=sample1.columns
col2=sample2.columns
cols=list(set(col1).intersection(set(col2)))
cols.sort()
data1=sample1[cols]
data2=sample2[cols]
```

```python
print(data2.shape)
data1.head()
```

```python
print(data2.shape)
data2.head()
```

```python
print('Select the columns to join: \nNote: The selected columns in an order if grouped should make the record level unique')
for i in range(len(cols)):
    print(str(i+1)+' : '+cols[i])
col_sep=input('\nEnter the columns separated by comma: ')
joining_cols=[cols[int(i)-1] for i in col_sep.split(',')]
```

```python
df=data1.set_index(joining_cols).join(data2.set_index(joining_cols),how='outer', lsuffix='_1', rsuffix='_2')
```

```python
from datetime import datetime
time=datetime.now()
file='rec'+'_'+time.strftime("%j")+time.strftime("%Y")+time.strftime("%H")+time.strftime("%M")+time.strftime("%S")+'.csv'
file_path='C:\\Users\\sselvaku\\OneDrive - Systech Solutions, Inc\Dev\\'+file
```

```python
status_dict=dict()
for i in cols:
    if i not in joining_cols:
        if all(df[i+'_1']==df[i+'_2'])==True:
            status_dict[i]=True
        else:
            status_dict[i]=False
```

```python
if all(status_dict.values())==True:
    print('Records are matching')
else:
    mis_index=set()
    mis_cols=[]
    for i in status_dict.keys():
        if status_dict[i]==False:
            mis_cols.append(i)
            for j in range(df.shape[0]):
                if df[i+'_1'].iloc[j]!=df[i+'_2'].iloc[j]:
                    mis_index.add(j)
    un_df=df.iloc[list(mis_index)]
    print(str(len(mis_index))+' records unmatched')
    print('Column(s): '+', '.join(mis_cols))
    out=input('Do you want the unmatched records to be stored locally (y/n): ')
    if out=='y' or out=='Y':
        un_df.to_csv(file_path)
        print('File saved successfully '+file_path)
    else:
        print('Thank You')
```
